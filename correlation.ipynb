```python
import numpy as np
import pandas as pd
import pingouin as pg
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
import math
import hyppo
from matplotlib import pyplot as plt
```

## –ó–∞–¥–∞—á–∞ 1

```python
def f1(x):
    f=[2*k+5 for k in x]
    return f

def f2(x):
    f=[k**2 for k in x]
    return f

def f3(x):
    f=[k*np.cos(k) for k in x]
    return f
```

```python
x1=np.random.uniform(0,1,100)
x2=np.random.uniform(-10,10,100)
u=np.random.uniform(-2,2,100)
```

```python
y_1 = [f1(x1)+u, f1(x2)+u]

y_2 = [f2(x1)+u, f2(x2)+u]

y_3 = [f3(x1)+u, f3(x2)+u]

x = [x1, x2]
```

```python
print("p-value, —á—Ç–æ –•=R[0,1] –∏–ª–∏ X=R[-10, 10] –∏ Y=2—Ö+5+U –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã: \n")
for i in range(len(y_1)): 
    test_pearson=st.pearsonr(x[i],y_1[i])
    test_spearman=st.spearmanr(x[i],y_1[i])
    test_kendall=st.kendalltau(x[i],y_1[i])
    mgc=hyppo.independence.MGC()
    test_mgc=mgc.test(x[i],y_1[i]) 
    print("–ü–∏—Ä—Å–æ–Ω:", test_pearson[1])
    print("–°–ø–∏—Ä–º–∞–Ω:", test_spearman.pvalue)
    print("–ö–µ–Ω–¥–∞–ª–ª:", test_kendall.pvalue)
    print("MGC:", test_mgc.pvalue)
    print('_' * 50)
```

```python
print("p-value, —á—Ç–æ –•=R[0,1] –∏–ª–∏ –•=R[-10,10] –∏ Y=x^2 –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã: \n")
for i in range(len(y_2)): 
    test_pearson=st.pearsonr(x[i],y_2[i])
    test_spearman=st.spearmanr(x[i],y_2[i])
    test_kendall=st.kendalltau(x[i],y_2[i])
    mgc=hyppo.independence.MGC()
    test_mgc=mgc.test(x[i],y_2[i]) #needs too much time...but still work, i'm surprised
    print("–ü–∏—Ä—Å–æ–Ω:", test_pearson[1])
    print("–°–ø–∏—Ä–º–∞–Ω:", test_spearman.pvalue)
    print("–ö–µ–Ω–¥–∞–ª–ª:", test_kendall.pvalue)
    print("MGC:", test_mgc.pvalue)
    print('_' * 50)
```

## –ó–∞–¥–∞—á–∞ 2

–£—Å—Ç—Ä–æ–π—Ç–µ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –∫–æ–ª—å—Ü–µ 1/2 < ùëã^2 + ùëå^2 < 1 (–≤–æ–∑—å–º–∏—Ç–µ –≤—ã–±–æ—Ä–∫–∏ –æ–±—ä–µ–º–∞ 100 –∏ 50). –ò—Å—Å–ª–µ–¥—É–π—Ç–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É ùëã –∏ ùëå —Å –ø–æ–º–æ—â—å—é –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –ö–µ–Ω–¥–µ–ª–ª–∞, –ø–æ–¥—Ö–æ–¥–∞ MGC –∏ –ø–æ–¥—Ö–æ–¥–∞ HHG. –¢–µ–ø–µ—Ä—å —Ç–æ –∂–µ —Å–∞–º–æ–µ –¥–ª—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–µ ‚àí1 ‚â§ ùëã ‚â§ 1, ‚àí1 ‚â§ ùëå ‚â§ 1. –¢–µ–ø–µ—Ä—å —Ç–æ –∂–µ —Å–∞–º–æ–µ –Ω–∞ –º–Ω–æ–∂–µ—Å—Ç–≤–µ, —Å–æ—Å—Ç–æ—è—â–µ–º –∏–∑ —á–µ—Ç—ã—Ä–µ—Ö –∫—Ä—É–≥–æ–≤ —Å —Ü–µ–Ω—Ç—Ä–∞–º–∏ –≤ (¬±1, ¬±1) –∏ —Ä–∞–¥–∏—É—Å–∞ 1/2.

```python
points = dict()
sizes = [50, 100, 1000]
for n in sizes: 
    points[n] = {'x': np.array([]), 'y': np.array([])}
    i = 0
    while i < n: 
        x = st.uniform.rvs(-1, 2, size=1)
        y = st.uniform.rvs(-1, 2, size=1)
        if 0.5 <= x**2 + y**2 <= 1: 
            points[n]['x'] = np.concatenate([points[n]['x'], x])
            points[n]['y'] = np.concatenate([points[n]['y'], y])
            i += 1
```

```python
# –Ω–∞ –º–Ω–æ–∂–µ—Å—Ç–≤–µ, —Å–æ—Å—Ç–æ—è—â–µ–º –∏–∑ —á–µ—Ç—ã—Ä–µ—Ö –∫—Ä—É–≥–æ–≤ —Å —Ü–µ–Ω—Ç—Ä–∞–º–∏ –≤ (¬±1, ¬±1) –∏ —Ä–∞–¥–∏—É—Å–∞ 1/2.
circles = dict()
sizes = [50, 100, 1000]
for n in sizes: 
    circles[n] = {'x': np.array([]), 'y': np.array([])}
    i = 0
    while i < n: 
        x = st.uniform.rvs(-2, 4, size=1)
        y = st.uniform.rvs(-2, 4, size=1)
        if (x-1)**2 + (y-1)**2 <= 1/4 or (x+1)**2 + (y-1)**2 <= 1/4 or (x-1)**2 + (y+1)**2 <= 1/4 or (x+1)**2 + (y+1)**2 <= 1/4:
            circles[n]['x'] = np.concatenate([circles[n]['x'], x])
            circles[n]['y'] = np.concatenate([circles[n]['y'], y])
            i += 1
```

```python
plt.figure(figsize=(10,5), dpi=200)
plt.plot(circles[1000]['x'], circles[1000]['y'], 'o')
plt.show()
```

```python
matrix = np.array([[2, 0.5, 1], [0.4, -1, 0.4], [0, 0.5, 1]])
# —Å–æ–∑–¥–∞–Ω–∏–µ 3 —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π x, y, z
data = np.array([np.random.exponential(1, 100) for item in ['x', 'y', 'z']])
df = pd.DataFrame(data=data.T, columns=['x','y','z'])

# —Å–æ–∑–¥–∞–Ω–∏–µ –µ—â–µ 3 —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π u, v, w, –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –ª–∏–Ω–µ–π–Ω—ã–º –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ–º –º–∞—Ç—Ä–∏—Ü—ã x, y, z
df1 = pd.DataFrame(data=df.values@matrix, columns=['u', 'v', 'w'])
df = df.join(df1)
corr = df.corr()
pcorr = df.pcorr()

sns.set(font_scale = 0.5)

plt.figure(figsize=(4, 3), dpi=300)
heatmap = sns.heatmap(corr, vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':9}, pad=12);

plt.figure(figsize=(4, 3), dpi=300)
heatmap = sns.heatmap(pcorr, vmin=-1, vmax=1, annot=True)
heatmap.set_title('Partial correlation Heatmap', fontdict={'fontsize':9}, pad=12);
```
