## Задача 1

```python
def f1(x):
    f=[2*k+5 for k in x]
    return f

def f2(x):
    f=[k**2 for k in x]
    return f

def f3(x):
    f=[k*np.cos(k) for k in x]
    return f
```

```python
x1=np.random.uniform(0,1,100)
x2=np.random.uniform(-10,10,100)
u=np.random.uniform(-2,2,100)
```

```python
y_1 = [f1(x1)+u, f1(x2)+u]

y_2 = [f2(x1)+u, f2(x2)+u]

y_3 = [f3(x1)+u, f3(x2)+u]

x = [x1, x2]
```

```python
print("p-value, что Х=R[0,1] или X=R[-10, 10] и Y=2х+5+U независимы: \n")
for i in range(len(y_1)): 
    test_pearson=st.pearsonr(x[i],y_1[i])
    test_spearman=st.spearmanr(x[i],y_1[i])
    test_kendall=st.kendalltau(x[i],y_1[i])
    mgc=hyppo.independence.MGC()
    test_mgc=mgc.test(x[i],y_1[i]) 
    print("Пирсон:", test_pearson[1])
    print("Спирман:", test_spearman.pvalue)
    print("Кендалл:", test_kendall.pvalue)
    print("MGC:", test_mgc.pvalue)
    print('_' * 50)
```

```python
print("p-value, что Х=R[0,1] или Х=R[-10,10] и Y=x^2 независимы: \n")
for i in range(len(y_2)): 
    test_pearson=st.pearsonr(x[i],y_2[i])
    test_spearman=st.spearmanr(x[i],y_2[i])
    test_kendall=st.kendalltau(x[i],y_2[i])
    mgc=hyppo.independence.MGC()
    test_mgc=mgc.test(x[i],y_2[i]) #needs too much time...but still work, i'm surprised
    print("Пирсон:", test_pearson[1])
    print("Спирман:", test_spearman.pvalue)
    print("Кендалл:", test_kendall.pvalue)
    print("MGC:", test_mgc.pvalue)
    print('_' * 50)
```

## Задача 2

Устройте равномерное распределение на кольце 1/2 < 𝑋^2 + 𝑌^2 < 1 (возьмите выборки объема 100 и 50). Исследуйте зависимость между 𝑋 и 𝑌 с помощью коэффициента Кенделла, подхода MGC и подхода HHG. Теперь то же самое для равномерного распределения на квадрате −1 ≤ 𝑋 ≤ 1, −1 ≤ 𝑌 ≤ 1. Теперь то же самое на множестве, состоящем из четырех кругов с центрами в (±1, ±1) и радиуса 1/2.

```python
points = dict()
sizes = [50, 100, 1000]
for n in sizes: 
    points[n] = {'x': np.array([]), 'y': np.array([])}
    i = 0
    while i < n: 
        x = st.uniform.rvs(-1, 2, size=1)
        y = st.uniform.rvs(-1, 2, size=1)
        if 0.5 <= x**2 + y**2 <= 1: 
            points[n]['x'] = np.concatenate([points[n]['x'], x])
            points[n]['y'] = np.concatenate([points[n]['y'], y])
            i += 1
```

```python
# на множестве, состоящем из четырех кругов с центрами в (±1, ±1) и радиуса 1/2.
circles = dict()
sizes = [50, 100, 1000]
for n in sizes: 
    circles[n] = {'x': np.array([]), 'y': np.array([])}
    i = 0
    while i < n: 
        x = st.uniform.rvs(-2, 4, size=1)
        y = st.uniform.rvs(-2, 4, size=1)
        if (x-1)**2 + (y-1)**2 <= 1/4 or (x+1)**2 + (y-1)**2 <= 1/4 or (x-1)**2 + (y+1)**2 <= 1/4 or (x+1)**2 + (y+1)**2 <= 1/4:
            circles[n]['x'] = np.concatenate([circles[n]['x'], x])
            circles[n]['y'] = np.concatenate([circles[n]['y'], y])
            i += 1
```

```python
plt.figure(figsize=(10,5), dpi=200)
plt.plot(circles[1000]['x'], circles[1000]['y'], 'o')
plt.show()
```

### Коэффициент Кенделла

```python
## кольцо
for size in sizes:
    print(f'Выборка размера {size}')
    test_kendall=st.kendalltau(points[size]['x'],points[size]['y'])
    print('Коэффициент Кенделла:', test_kendall[0])
    print('p_value для теста Кенделла:', test_kendall[1])
    print('_' * 50)
```

```python
## 4 круга
for size in sizes:
    print(f'Выборка размера {size}')
    test_kendall=st.kendalltau(circles[size]['x'],circles[size]['y'])
    print('Коэффициент Кенделла:', test_kendall[0])
    print('p_value для теста Кенделла:', test_kendall[1])
    print('_' * 50)
```

### Подход MGC

**кольцо**

```python
mgc=hyppo.independence.MGC()

for size in [50, 100]:
    print(f'Выборка размера {size}')
    test = mgc.test(points[size]['x'],points[size]['y'])
    print('Коэффициент MGC:', test[0])
    print('p_value для теста MGC:', test[1])
    print('_' * 50)
```

**4 круга**

```python
for size in [50, 100]:
    print(f'Выборка размера {size}')
    test = mgc.test(circles[size]['x'],circles[size]['y'])
    print('Коэффициент MGC:', test[0])
    print('p_value для теста MGC:', test[1])
    print('_' * 50)
```

### Подход HHG.

**кольцо**

```python
hhg=hyppo.independence.HHG()

for size in [50, 100]:
    print(f'Выборка размера {size}')
    test = hhg.test(points[size]['x'],points[size]['y'])
#     print('Статистика HHG:', test[0])
    print('p_value для теста HHG:', test[1])
    print('_' * 50)
```

```python
for size in [50, 100]:
    print(f'Выборка размера {size}')
    test = hhg.test(circles[size]['x'],circles[size]['y'])
#     print('Статистика HHG:', test[0])
    print('p_value для теста HHG:', test[1])
    print('_' * 50)
```

```python
matrix = np.array([[2, 0.5, 1], [0.4, -1, 0.4], [0, 0.5, 1]])
# создание 3 распределений x, y, z
data = np.array([np.random.exponential(1, 100) for item in ['x', 'y', 'z']])
df = pd.DataFrame(data=data.T, columns=['x','y','z'])

# создание еще 3 распределений u, v, w, полученных линейным преобразованием матрицы x, y, z
df1 = pd.DataFrame(data=df.values@matrix, columns=['u', 'v', 'w'])
df = df.join(df1)
corr = df.corr()
pcorr = df.pcorr()

sns.set(font_scale = 0.5)

plt.figure(figsize=(4, 3), dpi=300)
heatmap = sns.heatmap(corr, vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':9}, pad=12);

plt.figure(figsize=(4, 3), dpi=300)
heatmap = sns.heatmap(pcorr, vmin=-1, vmax=1, annot=True)
heatmap.set_title('Partial correlation Heatmap', fontdict={'fontsize':9}, pad=12);
```

Можно заметить, что у коэффициентов корреляции поменялись знаки при переходе к partial correlation matrix. Посмотрим на x и u, в первом случае корреляция практически 1, но при устранении влияния других факторов, она стала отрицательной, значит в первом случае мы присутствовало сильное влияние на факторы x и u.

```python
sns.set(font_scale = 2)
ax = sns.pairplot(df, height= 5 , aspect= 1)
ax.figure.subplots_adjust(top=0.96)
ax.fig.suptitle("The diagramm of joint distributions", fontsize=40)
plt.show(ax)
```
